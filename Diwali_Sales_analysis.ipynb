**Exploratory Data Analysis:**

```python
#Which gender spent more on average during Diwali?

var=df.groupby('Gender')['Amount'].mean()
var
```

```python
plt.bar(var.index,var.values,color='lightblue')
plt.title('Bar chart Male vs Female average purchase amount.')
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.show()
```

```python
#Which age group contributes the most to sales?

var=df.groupby('Age')['Amount'].sum()
var
```

```python
plt.bar(var.index,var.values,color='blue')
plt.title('Age group vs Total purchase amount.')
plt.xlabel('Age Group')
plt.ylabel('Amount')
plt.show()
```

```python
#Which marital status (Married / Unmarried) spent more overall?

var=df.groupby('Marital_Status')['Amount'].sum()
var
```

```python
plt.bar(var.index,var.values,color='blue')
plt.title('Marital_Status and plot bar chart.')
plt.xlabel('Marital_Status')
plt.ylabel('Amount')
plt.xticks([0,1],['Married','Unmarried'])
plt.show()
```

```python
#What’s the spending pattern based on occupation?
var1=df.groupby('Occupation')['Amount'].sum()
var1
```

```python
plt.barh(var1.index,var1.values,color='blue')
plt.title('Occupation vs Total Purchase.')
plt.xlabel('Amount')
plt.ylabel('Occupation')
plt.show()
```

```python
#Which state has the highest sales and number of orders?

var2=df.groupby('State')[['Amount','Orders']].sum().sort_values(by='Amount',ascending=False).head(10)
var2
```

```python
plt.barh(var2.index,var2['Amount'],color='blue')
plt.title(' Top 10 State by Sales')
plt.xlabel('Amount')
plt.ylabel('State')
plt.show()
```

```python
#Which product category generated the highest revenue?

df['Revenue']=df['Orders']*df['Amount']
df
```

```python
var3=df.groupby('Product_Category')['Revenue'].sum()
var3
```

```python
plt.barh(var3.index,var3.values,color='blue')
plt.title('Revenue by Product category')
plt.xlabel('Revenue')
plt.ylabel('Product Category')
plt.show()
```

```python
#Top 10 most sold products during Diwali?

var4=df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=False).head(10)
var4
```

```python
plt.barh(var4.index,var4.values,color='blue')
plt.title('Top 10 most sold products during Diwali ')
plt.xlabel('Product Count')
plt.ylabel('Product Category')
plt.show()
```

```python
#Average purchase amount by age group for each product category

result = df.groupby(["Age Group","Product_Category"])["Amount"].mean().reset_index()
print(result.head())
```

```python
plt.bar(result["Age Group"],result["Amount"])
plt.xlabel("Age Group")
plt.ylabel("Average Purchase Amount")
plt.title("Average purchase amount by age group")
plt.show()
```

```python
# Which product category is more popular among men vs women?

result = df.groupby(["Product_Category","Gender"])["Orders"].sum().reset_index()
pivot_df = result.pivot(index="Product_Category", columns="Gender", values="Orders")
print(pivot_df)
```

```python
pivot_df.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Product Category Popularity by Gender (Stacked)")
plt.ylabel("Number of Orders")
plt.show()
```
